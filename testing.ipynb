```python
import pandas as pd

df = pd.DataFrame({ "country":["abc, USA","USA","HK","United States","China","US"]})

df
                   
```

```python
#testing1 the .str.contains() and findall() to search for sunstrings in Pandas column

df.loc[df.country.str.contains("USA|United States", na=False, case=False),"New"] = "United States"
df
```

```python
#testing2 using a loop to correct multiple data

def correction(list_name):
    for element in list_name:
        df.loc[df.country.str.contains(element, na=False, case=False),"New2"] = list_name[0]
US = ["USA","United States","US"]
HK = ["HK"]

#put the lists into a list

all = [US,HK]
for each in all:
    correction(each)

df
```
